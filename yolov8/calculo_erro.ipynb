```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

path_dados = "dados/percurso1/labels/*.txt"
path_fotos = "dados/percurso1/*.jpg"
path_save_imgs = "resultados/*.jpg"

escala = 1280  # em pixels
escala = 720  # em pixels

lagura_caminho = 1000  # em cm

erro = [0] * 7

# classe 0 = car
# classe 1 = path
```

```python
# classe 0 = car
# classe 1 = path

n = 0
path_data = path_dados.replace('*', str(n))
path_img = path_fotos.replace('*', str(n))
path_save_img = path_save_imgs.replace('*', str(n))

coordenadas = {'0':[], '1':[], 'formato':str}

with open(path_data, 'r') as f:
    for c in f:
        a = c.removesuffix('\n').split()
        classe = a[0]
        coordenadas[classe].append(list(map(lambda x: float(x), a[1:])))
        
        coordenadas['formato'] = 'xyxyxyxyn' if len(a) > 5 else 'xywhn'

if coordenadas['formato'] == 'xywhn':
    x_correcao = 8 / 720
    y_correcao = 14 / 1280

    x_sup_left, y_sup_left = coordenadas['1'][0][0] - coordenadas['1'][0][2] / 2.0 + x_correcao, coordenadas['1'][0][1] - coordenadas['1'][0][3] / 2.0 + y_correcao
    x_sup_right, y_sup_right = coordenadas['1'][0][0] + coordenadas['1'][0][2] / 2.0 - x_correcao, coordenadas['1'][0][1] - coordenadas['1'][0][3] / 2.0
    x_target, y_target = (x_sup_left + x_sup_right) / 2.0, (y_sup_left + y_sup_right) / 2.0

    x_center_car, y_center_car = coordenadas['0'][0][:2]

    escala = lagura_caminho / ((x_sup_right - x_sup_left) ** 2 + (y_sup_left - y_sup_right) ** 2 ) ** (1 / 2)

    x_erro, y_erro = escala * (x_target - x_center_car), escala * (y_target - y_center_car)

    erro[n] = x_erro ** 2 + y_erro ** 2
    
    #plt.plot([b * 1280 for b in [x_sup_left, x_sup_right, x_target, x_center_car]], [b * 720 for b in [y_sup_left, y_sup_right, y_target, y_center_car]], c='r')
    plt.plot([b * 1280 for b in [x_target, x_center_car]], [b * 720 for b in [y_target, y_center_car]], c='r')

img = mpimg.imread(path_img)
imgplot = plt.imshow(img)
plt.savefig(path_save_img, dpi=480)
plt.show()
```
