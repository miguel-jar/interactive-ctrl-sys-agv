```python

from ultralytics import YOLO

import shutil
import os

# Select current PATH for testing
PATH = 'path_2'
current_path = f"imgs/{PATH}/*.jpg"
```

```python
YOLO_MODEL = 'obb_v3'

if os.path.exists(f'data/{PATH}/{YOLO_MODEL}'):
    shutil.rmtree(f'data/{PATH}/{YOLO_MODEL}')
    print("Folder sucessfully removed.")

model = YOLO(f'models/{YOLO_MODEL}/best.pt')
results = model.predict(current_path, imgsz=(736, 736), show=False, save=True, iou=0.4, save_txt=True, save_conf=True, project='data', name=f'{PATH}/{YOLO_MODEL}')
```
