```python
from ultralytics import YOLO

import shutil
import os

# Select current PATH for testing
ROBOT_PATH = 'path_1'

robot_path_imgs = f"imgs/{ROBOT_PATH}/*.jpg"
project_path = os.path.abspath('data')
```

```python
YOLO_MODEL = 'v3'

if os.path.exists(f'data/{ROBOT_PATH}/{YOLO_MODEL}'):
    shutil.rmtree(f'data/{ROBOT_PATH}/{YOLO_MODEL}')
    print("Folder sucessfully removed.")

model = YOLO(f'models/{YOLO_MODEL}/best.pt')
results = model.predict(robot_path_imgs, imgsz=(736, 736), show=False, save=True, iou=0.4, save_txt=True, save_conf=True, project=project_path, name=f'{ROBOT_PATH}/{YOLO_MODEL}')
```
